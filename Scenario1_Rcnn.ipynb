# Selective Search

```python
import cv2
import json
```

```python
cv2.setUseOptimized(True);
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
```

```python
im = cv2.imread(os.path.join(train_path_pill,"image","VAIPE_P_0_0.jpg"))

scale_percent = 10 # percent of original size
width = int(im.shape[1] * scale_percent / 100)
height = int(im.shape[0] * scale_percent / 100)
dim = (width, height)

im = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)
ss.setBaseImage(im)
ss.switchToSelectiveSearchFast()
rects = ss.process()
imOut = im.copy()
imOutLabel =im.copy()
###
height = 1
width = 2

f, axarr = plt.subplots(height, width) 
f.set_figheight(height * 5)
f.set_figwidth(width * 5)
#####
label = os.path.join(train_path_pill,"label", "VAIPE_P_0_0.json")
#  print(label)
with open(label) as f:
  data = json.load(f)
for datat in data:
  x1 = int(datat["x"] * scale_percent / 100)
  y1 = int(datat["y"] * scale_percent / 100)
  x2 = int((datat["x"]+datat["w"])* scale_percent / 100)
  y2 = int((datat["y"]+datat["h"])* scale_percent / 100)
  cv2.rectangle(imOutLabel, (x1, y1),
              (x2, y2),
              (0, 255, 0), 1, cv2.LINE_AA)

##############

for i, rect in (enumerate(rects)):
    x, y, w, h = rect
#     print(x,y,w,h)
#     imOut = imOut[x:x+w,y:y+h]
    cv2.rectangle(imOut, (x, y), (x+w, y+h), (0, 255, 0), 1, cv2.LINE_AA)
# plt.figure()
axarr[0].imshow(imOut)
axarr[1].imshow(imOutLabel)
```

```python
def get_iou(bb1, bb2):
  # assuring for proper dimension.
    assert bb1['x1'] < bb1['x2']
    assert bb1['y1'] < bb1['y2']
    assert bb2['x1'] < bb2['x2']
    assert bb2['y1'] < bb2['y2']
  # calculating dimension of common area between these two boxes.
    x_left = max(bb1['x1'], bb2['x1'])
    y_top = max(bb1['y1'], bb2['y1'])
    x_right = min(bb1['x2'], bb2['x2'])
    y_bottom = min(bb1['y2'], bb2['y2'])
  # if there is no overlap output 0 as intersection area is zero.
    if x_right < x_left or y_bottom < y_top:
        return 0.0
  # calculating intersection area.
    intersection_area = (x_right - x_left) * (y_bottom - y_top)
  # individual areas of both these bounding boxes.
    bb1_area = (bb1['x2'] - bb1['x1']) * (bb1['y2'] - bb1['y1'])
    bb2_area = (bb2['x2'] - bb2['x1']) * (bb2['y2'] - bb2['y1'])
  # union area = area of bb1_+ area of bb2 - intersection of bb1 and bb2.
    iou = intersection_area / float(bb1_area + bb2_area - intersection_area)
    assert iou >= 0.0
    assert iou <= 1.0
    return iou
```

# Prepare training data

```python
# At the end of below code we will have our train data in these lists
train_images=[]
train_labels=[]
```

```python
listimages = train_pill_fnames[0:200]
listlabels = train_pill_labels[0:200]
```

```python
listresults = []
for e,i in enumerate(listimages):
##############image
  image = cv2.imread(os.path.join(train_path_pill,"image",i))
#  print(os.path.join(train_path_pill,"image",i))
#resize
  scale_percent = 10 # percent of original size
  width = int(image.shape[1] * scale_percent / 100)
  height = int(image.shape[0] * scale_percent / 100)
  dim = (width, height)

  im = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

  ss.setBaseImage(im)
  ss.switchToSelectiveSearchFast()
  ssresults = ss.process()

  imout = im.copy()
############label
  label = os.path.join(train_path_pill,"label", listlabels[e])
#  print(label)
  with open(label) as f:
    data = json.load(f)

  gtvalues=[]
  for datat in data:
     gtvalues.append({"x1":int(datat["x"] * scale_percent /100),
                      "x2":int((datat["x"]+datat["w"]) * scale_percent /100),
                      "y1":int(datat["y"] * scale_percent /100),
                      "y2":int((datat["y"]+datat["h"]) * scale_percent /100),
                      "label":int(datat["label"]) })
  counter = 0
  falsecounter = 0
  flag = 0
  fflag = 0
  bflag = 0
  for e,result in enumerate(ssresults):
    if e < 2000 and flag == 0:     # till 2000 to get top 2000 regions only
      for gtval in gtvalues:
        x,y,w,h = result
        iou = get_iou({"x1":gtval["x1"],
                     "x2":gtval["x2"],
                     "y1":gtval["y1"],
                     "y2":gtval["y2"]},
                    {"x1":x,"x2":x+w,"y1":y,"y2":y+h})  # calculating IoU for each of the proposed regions
        print(iou)
        if counter < 30:       # getting only 30 psoitive examples
          if iou > 0.70:     # IoU or being positive is 0.7
            timage = imout[x:x+w,y:y+h]
            try:
              resized = cv2.resize(timage, (224,224), interpolation = cv2.INTER_AREA)
            except:
              break
            train_images.append(resized)
            train_labels.append(gtval["label"])
            counter += 1
        else :
          fflag =1              # to insure we have collected all psotive examples
        if falsecounter <5:      # 30 negatve examples are allowed only
          if iou < 0.3:         # IoU or being negative is 0.3
            timage = imout[x:x+w,y:y+h]
            try: 
              resized = cv2.resize(timage, (224,224), interpolation = cv2.INTER_AREA)
            except:
              break
            train_images.append(resized)
            train_labels.append(0)
            falsecounter += 1
        else :
          bflag = 1             #to ensure we have collected all negative examples
      if fflag == 1 and bflag == 1:  
        flag = 1        # to signal the complition of data extaction from a particular image
```

# Model

```python
import tensorflow as tf
from keras.layers import Dense
from keras import Model
from keras import optimizers
```

```python
vgg = tf.keras.applications.vgg16.VGG16(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
for layer in vgg.layers[:-2]:
  layer.trainable = False
x = vgg.get_layer('fc2')
last_output =  x.output
last_output = tf.keras.layers.Dropout(0.4)(last_output) 

x = tf.keras.layers.Dense(1171,activation = 'softmax')(last_output)  
model = tf.keras.Model(vgg.input,x)
model.compile(optimizer = "adam", 
              loss = 'sparse_categorical_crossentropy', 
              metrics = ['acc'])
```

```python
model.summary()
```

```python
hist_final = model.fit(X_new,Y_new,batch_size = 20,epochs = 5, verbose = 1,validation_split=.1,shuffle = True)
```

```python
import matplotlib.pyplot as plt
# plt.plot(hist.history["acc"])
# plt.plot(hist.history['val_acc'])
plt.plot(hist_final.history['loss'])
plt.plot(hist_final.history['val_loss'])
plt.title("model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["Loss","Validation Loss"])
plt.show()
plt.savefig('chart loss.png')
```
